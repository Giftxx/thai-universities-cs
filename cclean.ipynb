```python
import pandas as pd
import re

# โหลดไฟล์ Excel
file_path = "tcas_engineering_programs.xlsx"
df = pd.read_excel(file_path, sheet_name="Processing_Data")

# ฟังก์ชันดึงค่าจากข้อความค่าใช้จ่าย
def extract_tuition(text):
    if pd.isna(text):
        return None, None
    text = str(text)

    # Case 1: ภาคการศึกษาละ ...
    match1 = re.search(r'ภาคการศึกษาละ\s*([\d,]+)', text)
    if match1:
        return int(match1.group(1).replace(',', '')), 'per_term'

    # Case 2: xxx/ภาค หรือ ต่อภาคการศึกษา
    match2 = re.search(r'([,\d]+)[^\d]{0,5}(?:บาท)?\s*(?:/|ต่อ)?\s*ภาค', text)
    if match2:
        return int(match2.group(1).replace(',', '')), 'per_term'

    # Case 3: ตัวเลขล้วน น่าจะเป็นแบบรวม
    match3 = re.search(r'([\d,]+)', text)
    if match3:
        value = int(match3.group(1).replace(',', ''))
        if value >= 100000:
            return value, 'lumpsum'

    return None, None

# ประมวลผล
df[['ค่าใช้จ่าย_ตัวเลข', 'ประเภทค่าใช้จ่าย']] = df['ค่าใช้จ่าย'].apply(
    lambda x: pd.Series(extract_tuition(x))
)

# คำนวณค่าเทอมต่อภาค
df['ค่าเทอมต่อเทอม_ประมาณ'] = df.apply(
    lambda row: row['ค่าใช้จ่าย_ตัวเลข'] / 8 if row['ประเภทค่าใช้จ่าย'] == 'lumpsum'
    else row['ค่าใช้จ่าย_ตัวเลข'], axis=1
)

# บันทึกไฟล์ผลลัพธ์
df.to_excel("cleaned_tuition_data_v2.xlsx", index=False)

# ตัวอย่างผลลัพธ์
# print(df[['หลักสูตร', 'ค่าใช้จ่าย', 'ประเภทค่าใช้จ่าย', 'ค่าเทอมต่อเทอม_ประมาณ']].head())
df
```
